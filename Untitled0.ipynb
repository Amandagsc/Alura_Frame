```python
# Calcular quantos dias um produto duraria se apessoa usar X porções por dia



quantidade_produto = int(input("Digite a quantidade do produto que vem na embalagem: "))
quantidade_uso_diario =  int(input("digite quantas porções do produto será usado por dia (gramas/ml): "))
dias = int(quantidade_produto / quantidade_uso_diario)
print(f'O produto vai durar {int(dias)} dias')
```
